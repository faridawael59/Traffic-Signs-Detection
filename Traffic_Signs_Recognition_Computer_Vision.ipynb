```python
#this file is used for  yolo to tell it the paths of the data, i corrected them to the correct paths

yaml_path = "/content/car/data.yaml"

with open(yaml_path, 'r') as file:
    data = file.read()

# Replace relative paths with absolute ones
data = data.replace('../train/images', '/content/car/train/images')
data = data.replace('../valid/images', '/content/car/valid/images')
data = data.replace('../test/images', '/content/car/test/images')

# Save the updated file
with open(yaml_path, 'w') as file:
    file.write(data)

print("✅ data.yaml updated")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')

model.train(
    data='/content/car/data.yaml',
    epochs=20,                 # Train longer for better results
    imgsz=416,  # Set image size to 416x416 pixels to standardize inputs for the model; smaller sizes train faster and use less memory
    batch=16,   # Number of images processed at once before updating model weights; balances speed and memory usage
    name='traffic-sign-v8'
)
```

```python
metrics = model.val() #prints metrics
```

```python
results = model.predict(source='/content/car/test/images', save=True, conf=0.25) #Sets the confidence threshold — only predictions with confidence ≥ 0.25 will be kept.

```

```python
from ultralytics import YOLO
import shutil
import os

model = YOLO('/content/yolov8s.pt')
model.predict(
    source='/content/video.mp4',  # Replace with your actual video file path
    save=True,                             # Saves the output video with detections
    conf=0.25,                             # Confidence threshold
    imgsz=416                              # Image size used during training
)

# Find the last predict folder
predict_dir = 'runs/detect/predict'
output_video = os.path.join(predict_dir, 'video.avi')  # Default YOLO output

# Define your custom path
custom_path = '/content/traffic_sign_result2.mp4'

# Create output directory if it doesn't exist
os.makedirs(os.path.dirname(custom_path), exist_ok=True)

# Copy the result video to your desired location
shutil.copy(output_video, custom_path)

print(f"✅ Saved prediction result to: {custom_path}")
```
